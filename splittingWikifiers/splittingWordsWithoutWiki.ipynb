```python
import os
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from collections import defaultdict
```

```python
def tokenize(directory_input, input_file, tokens_occ):
   with open(os.path.join(directory_input, input_file), "rt", encoding="utf-8") as f:
       head = [next(f) for x in range(4500)]   
       for word in head:
           sep = ' '
           rest = word.split(sep, 1)[0]
           #print(rest)
           tokens_occ[rest] += 1
   return tokens_occ
```

```python
tokens_occ = defaultdict(lambda: 0)
```

```python
tokens_occ = tokenize("C:/Users/shwet/ProjectFinal/corpusTokenization/", "all_wiki_words_sorted_per_occ.txt", tokens_occ)
```

```python
def writeSplittedWikiFiles(oneword_vocab, multipleword_vocab, tokens_occ):
    oneword_file = open(oneword_vocab, 'w', encoding="utf-8")
    multipleword_file = open(multipleword_vocab, 'w', encoding="utf-8")
    for word in tokens_occ:
        print ((word))
        replacedWord = word.replace('Wiki__', '')
        lower = replacedWord.lower()
        if(lower.find("_")):
            withoutWikifiers=lower.split("_")
        while("" in withoutWikifiers) : 
            withoutWikifiers.remove("")
        withoutWikifiers.pop()
        totalWords = len(withoutWikifiers)
        for n in withoutWikifiers:
            if totalWords == 1:
                 oneword_file.write(str("%s\n" %n))
            else:
                 multipleword_file.write(str("%s\n" %n))
```
